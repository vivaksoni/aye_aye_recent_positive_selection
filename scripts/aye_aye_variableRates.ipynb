```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy import stats
import seaborn as sns
%matplotlib inline
import scipy as sp
# import statsmodels.api as sm
# import statsmodels.formula.api as smf
import os
import glob

import itertools

import string
#from Bio import AlignIO
```

```python
totalWins = 92
reps = 100
windowSize = 1000
```

```python
for i in range(1, 101):
    rrm = 5
    #Chuck out any replicates whose mean is too far from the fixed mean, using CIs as thresholds
    while((rrm>1.57e-8) | (rrm<1.47e-8)):
        pos = [int(x+1) for x in range(0, totalWins*windowSize, windowSize)]

        #Generate random variables (a and b are 1.52 multiplied and divided by 3 respectively)
        mu = stats.loguniform(a= 0.61e-8, 
                  b=3.8e-8).rvs(totalWins)
        rdf = pd.DataFrame([pos, mu]).T
        rdf[0] = [int(x) for x in rdf[0]]
        rrm = rdf[1].mean()
    print(str(rrm))
    rdf.to_csv(r"/home/vivak/aye_aye/mu_maps/"+str(i)+".txt", 
              sep='\t', 
              header=False, 
              index=False)
```

```python
for i in range(1, 101):
    rrm = 3
    #Chuck out any replicates whose mean is too far from the fixed mean
    while((rrm>1.25) | (rrm<0.75)):
        pos = [int(x+1) for x in range(0, totalWins*windowSize, windowSize)]

        #Generate random variables (a is such that mean is close to Keightley et al. 2014 mean)
        mu = stats.loguniform(a= 0.01, 
                  b=10).rvs(totalWins)
        rdf = pd.DataFrame([pos, mu]).T
        rdf[0] = [int(x) for x in rdf[0]]
        rrm = rdf[1].mean()
    print(str(rrm))
    rdf.to_csv(r"/home/vivak/aye_aye/rr_maps/"+str(i)+".txt", 
              sep='\t', 
              header=False, 
              index=False)
```
